# Project: Daily Exchange Rates per Euro 1999-2021

The dataset describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union.

If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate).

Daria Chemkaeva put together the data set and made it available on Kaggle — the data source is the European Central Bank. Note that the dataset gets regular updates -I downloaded it on January 2022-.

```python
#Importing pandas library and exploring the first and the last 5 rows of the dataset

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style

exchange_rates = pd.read_csv('euro-daily-hist_1999_2021.csv')
exchange_rates.head(5)
```

```python
exchange_rates.tail(5)
```

```python
exchange_rates.info()
```

The dataset has 5877 rows and 45 columns. It starts from Jan 1999 and ends in Sep 2021. The columns represent various currncies. the data should be an integer; however it is represented as string (object). Also, the data has many null values inside many columns.

The main focus in this project is the exchange rate between the euro and the American dollar. So, the next step will be cleaning the data concerned with US Dollars.

## Cleaning Data

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['US_dollar', 'Time']].copy()
exchange_rates['US_dollar'].value_counts()
```

Drop all the rows where the (-) character (62 values) appears in the US_dollar column.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

After cleaning the data, the next step is to create a line plot represents the evolution of the euro-dollar exchange rate.

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

The small wiggles are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. It's better to focus on the long-term, so the next step will be calculating the rolling means for the US_dollar column using a moving window of 30 days, to work with the changing rates during different months.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar['rolling_mean'].value_counts()
```

## Part 2: The last three US Presidents Comparison:

The goal now is to show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)).

```python
# Parceling the data into 4 main dateranges:

euro_to_dollar['year'] = pd.DatetimeIndex(euro_to_dollar['Time']).year

bush_obama_trump = euro_to_dollar[(euro_to_dollar['year'] >= 2001) & (euro_to_dollar['year'] < 2021)]
bush = bush_obama_trump[bush_obama_trump['year'] < 2009]
obama = bush_obama_trump[(bush_obama_trump['year'] >= 2009) & (bush_obama_trump['year'] < 2017)]
trump = bush_obama_trump[(bush_obama_trump['year'] >= 2017) & (bush_obama_trump['year'] < 2021)]
```

```python
# Adding the FiveThirtyEight style
style.use('fivethirtyeight')

plt.figure(figsize=(12, 6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)

axes = [ax1, ax2, ax3, ax4]

# Making changes to all axes 

for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'],
                   alpha=0.3)
    ax.grid(alpha=0.5)    

# visualising Bush's period

ax1.plot(bush['Time'], bush['rolling_mean'],
         color='#BF5FFF')
ax1.text(731516.0, 1.92, 'BUSH', fontsize=18, weight='bold',
        color='#BF5FFF')
ax1.text(731216.0, 1.8, '2001-2009', 
        weight='bold', alpha=0.3)
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                     '2007', '', '2009'],
                   alpha=0.3)


# visualising Obama's period

ax2.plot(obama['Time'], obama['rolling_mean'],
         color='#ffa500')
ax2.text(734288.0, 1.92, 'OBAMA', fontsize=18, weight='bold',
        color='#ffa500')
ax2.text(734138.0, 1.8, '2009-2017', 
        weight='bold', alpha=0.3)
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                     '2015', '', '2017'],
                   alpha=0.3)

# visualising Trumps's period

ax3.plot(trump['Time'], trump['rolling_mean'],
         color='#00B2EE')
ax3.text(736855.0, 1.92, 'TRUMP', fontsize=18, weight='bold',
        color='#00B2EE')
ax3.text(736745.0, 1.8, '2009-2017', 
        weight='bold', alpha=0.3)
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '',
                     '2020', '', '2021'],
                   alpha=0.3)

# A comparison between the three prisedents

ax4.plot(bush['Time'], bush['rolling_mean'],
        color='#BF5FFF')
ax4.plot(obama['Time'], obama['rolling_mean'],
        color='#ffa500')
ax4.plot(trump['Time'], trump['rolling_mean'],
        color='#00B2EE')
ax4.grid(alpha=0.5)
ax4.set_xticks([])

#Adding a title and a subtitle
ax1.text(730016.0, 2.35, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(730016.0, 2.14, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''',
        fontsize=16)

# Adding a signature
ax4.text(729916.0, 0.65, '©MAHA FOUAD' + ' '*103 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)

plt.show()
```
